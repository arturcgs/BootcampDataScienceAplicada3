```python
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import matplotlib.dates as mdates
from pandas.plotting import autocorrelation_plot
import statsmodels.api as sm
from fbprophet import Prophet
from fbprophet.plot import add_changepoints_to_plot
```

```python
## Criando o df pra fazer o Machine Learning do Prophet

df = pd.DataFrame()

df['ds'] = dados_mococa['datahora']
df['y'] = dados_mococa['casos_novos']

df.head()
```

```python
#Treinando o modelo

df.shape #vendo qntos dados eu tenho, pra separar 70% pro modelo e 30% pro teste
         #eu acabei separando em 400, ao invés de 70%, pq o resultado deu melhor

df_treino = pd.DataFrame()
df_treino['ds'] = df['ds'][:400]
df_treino['y'] = df['y'][:400]

df_teste = pd.DataFrame()
df_teste['ds'] = df['ds'][400:]
df_teste['y'] = df['y'][400:]
```

```python
#Fazendo o modelo

modelo = Prophet()

modelo.fit(df_treino)
```

```python
#Fazendo a previsão pros 43 dias restantes

df_futuro = modelo.make_future_dataframe(periods = 43)
previsao = modelo.predict(df_futuro)
```

```python
#Fazendo o gráfico do teste

modelo.plot(previsao, figsize = (15, 8));
plt.plot(df_teste['ds'], df_teste['y'], '.r')
```

```python
#Gráfico de componentes

modelo.plot_components(previsao, figsize = (15, 8));
```

```python
#Criando função pra criar e rodar um modelo

def modelar (df : pd.DataFrame(), n_previsao : int, growth='linear', 
                   changepoints=None, n_changepoints=25, changepoint_range=0.8, 
                   yearly_seasonality='auto', weekly_seasonality='auto', 
                   daily_seasonality='auto', holidays=None, seasonality_mode='additive', 
                   seasonality_prior_scale=10.0, holidays_prior_scale=10.0, 
                   changepoint_prior_scale=0.05, mcmc_samples=0, interval_width=0.8, 
                   uncertainty_samples=1000, stan_backend=None):
  '''Função que recebe um Dataframe do Pandas para montar um modelo. 
  Também recebe um valor inteiro para o número de dias que fará a previsão.
  A função tbm aceita tds os parâmetros do Prophet'''

  modelo = Prophet(growth, changepoints, n_changepoints, changepoint_range, 
                  yearly_seasonality, weekly_seasonality, daily_seasonality, 
                  holidays, seasonality_mode, seasonality_prior_scale, 
                  holidays_prior_scale, changepoint_prior_scale, mcmc_samples, 
                  interval_width, uncertainty_samples, stan_backend)
  modelo.fit(df)

  df_futuro = modelo.make_future_dataframe(periods = n_previsao)

  previsao = modelo.predict(df_futuro)

  return modelo, previsao
```

```python
#Criando um modelo mudando o changepoint_prior_scale

modelo2, previsao2 = modelar(df_treino, 43, changepoint_prior_scale = 1, seasonality_mode='multiplicative')
```

```python
#Observando a mudança causada pelo changepoint_prior_scale

fig = modelo2.plot(previsao2, figsize=(15, 8))
a = add_changepoints_to_plot(fig.gca(), modelo2, previsao2)

plt.plot(df_teste['ds'], df_teste['y'], '.r')
```

```python
#Observando a mudança causada pelo changepoint_prior_scale

modelo2.plot(previsao2, figsize=(15, 8));

plt.plot(df_teste['ds'], df_teste['y'], '.-g')
```

```python
# Retirando os Outliers

novo_y = df_treino.reset_index()

sem_outliers = novo_y[(novo_y['y'] > previsao2['yhat_lower'][:400]) & (novo_y['y'] < previsao2['yhat_upper'][:400])]

df_sem_outliers = sem_outliers[['ds', 'y']]
df_sem_outliers.head()
```

```python
#Fazendo o modelo sem outliers
modelo_sem_outliers, previsao_sem_outliers = modelar(df_sem_outliers, 43, changepoint_prior_scale = 1, seasonality_mode='multiplicative')
```

```python
#Plotando o gráfico do modelo

modelo_sem_outliers.plot(previsao_sem_outliers, figsize=(15, 8));

plt.plot(df_teste['ds'], df_teste['y'], '.', color='orange')
```

```python
#Adicionando linhas pra avaliar o quaõ perto o modelo está dos dados

modelo_sem_outliers.plot(previsao_sem_outliers, figsize=(15, 8));

plt.plot(df_teste['ds'], df_teste['y'], '.-', color='green')

plt.plot(df_sem_outliers['ds'], df_sem_outliers['y'], '-', color='gray')

```

```python
import pandas as pd
```

```python
 #importando o banco de dados de gastos hospitalares, por data do procedimento
URL = 'https://raw.githubusercontent.com/arturcgs/BootcampDataScienceAplicada3/main/Bancos_de_Dados/Prod_Hosp_SUS%20-%20Atendimento.csv'
dados = pd.read_csv(URL, encoding="utf-8", 
                    skiprows = 3, sep=";", skipfooter=11, 
                    thousands=".", decimal=",")
dados.head()
```

```python
#Limpando o DataFrame

#pegando somente as colunas que tem tds os campos preenchidos
colunas_usaveis = dados.mean().index.tolist()
colunas_usaveis.insert(0, "Unidade da Federação") #inserindo o índice

dados_usavel = dados[colunas_usaveis]
dados_usavel = dados_usavel.set_index("Unidade da Federação") #adicionando o índice

#tirando a coluna e linha total
dados_usavel = dados_usavel.drop("Total", axis=1)
dados_usavel = dados_usavel.drop("Total", axis=0)

dados_usavel.head()
```

```python
#Importando a tabela de população por estado da Wikipedia
populacao_por_estado = pd.read_html('https://pt.wikipedia.org/wiki/Lista_de_unidades_federativas_do_Brasil_por_popula%C3%A7%C3%A3o')[0]

#Arrumando os nomes e tirando as colunas indesejadas
populacao_por_estado['uf'] = populacao_por_estado['Unidade federativa']
populacao_por_estado['populacao'] = populacao_por_estado['População']
populacao_por_estado = populacao_por_estado.drop(['Posição', 'Unidade federativa',	'População',	'% da pop. total',	'País comparável (habitantes)'], axis=1)

#tirando os espaços q tem entre os números
populacao_por_estado['populacao'] = populacao_por_estado['populacao'].str.replace(' ', '').astype(int)

#ordenando por ordem crescente
populacao_por_estado = populacao_por_estado.sort_values('populacao', ascending=False)

populacao_por_estado.head()
```
